```python
import pickle
import pandas as pd
import numpy as np
from IPython.display import display
pd.options.display.max_columns = None
pd.options.display.max_rows = None

pickle_in = open("match_dataframe.pickle","rb")
match_dataframe = pickle.load(pickle_in)

display(match_dataframe)

#Check Lobby Types
match_dataframe['LobbyType'].unique()

#Remove all rows with LobbyType 4, since these are bot matches
match_dataframe = match_dataframe[match_dataframe.LobbyType != 4]

#Check result
match_dataframe['LobbyType'].unique()

#Check Game Mode
match_dataframe['GameMode'].unique()

match_dataframe['GameMode'].value_counts()

#Since this account does not have any ranked matches (Game Mode #22), this means there is an error on the storing side. 
#So I am going to replace all 22 values with 1.
match_dataframe['GameMode'].replace(to_replace = 22, value = 1, inplace = True)

#Check result
match_dataframe['GameMode'].value_counts()

#Remove all values that are not 1
match_dataframe = match_dataframe[match_dataframe.GameMode == 1]

#Check result
match_dataframe['GameMode'].unique()

#Now that I have a certain kind of matches which I want to analyze, I can just remove the LobbyType and GameMode
#columns from the dataframe
del match_dataframe['LobbyType']
del match_dataframe['GameMode']

#Check result
display(match_dataframe)

#Check Leaver Status column
match_dataframe['LeaverStatus'].unique()

match_dataframe['LeaverStatus'].value_counts()

#Around 18% of my games have leavers, but since there is an equal chance of leavers being on either team,
#I can ignore this statistics and the 18% of my data in the dataframe.

#Store the result            
pickle_out = open("match_dataframe_processed.pickle","wb")
pickle.dump(match_dataframe, pickle_out)
pickle_out.close()

#Check if it saved
pickle_in = open("match_dataframe_processed.pickle","rb")
check_match_dataframe = pickle.load(pickle_in)

display(check_match_dataframe)
```
